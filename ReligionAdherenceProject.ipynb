```python
#importing libraries and the data 

import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
df=pd.read_csv(r"C:\Users\lenvo\Desktop\WRP national data.csv")
df.head()
```

DATA CLEANÄ°NG 

```python
#I am only interested to religions as whole , not sect by sect so i am adding every sect together as one column 
general_chrsts=df["chrstprot"]+df["chrstcat"]+df["chrstorth"]+df["chrstang"]+df["chrstothr"]
general_juds=df["judorth"]+df["jdcons"]+df["judref"]+df["judothr"]
general_islam=df["islmsun"]+df["islmshi"]+df["islmibd"]+df["islmnat"]+df["islmalw"]+df["islmahm"]+df["islmothr"]
general_budism=df["budmah"]+df["budthr"]+df["budothr"]
df["chrstprot"]=general_chrsts
df["chrstcat"]=general_juds
df["chrstorth"]=general_islam
df["chrstang"]=general_budism
df=df.rename(columns={"chrstprot":"christianity","chrstcat":"judaism","chrstorth":"islam","chrstang":"budism"})
df.head()


```

```python
#Dropping the columns that i don't need anymore
df.drop(df.columns[7:36], axis=1, inplace=True)
df.drop(df.columns[11:], axis=1, inplace=True)
df.drop(columns={"state","othrgen"},inplace=True)
```

```python
#I am trying create new data frame to get sum of religions only by years 
column_names=["year"]
df_new= pd.DataFrame(columns=column_names)
```

```python
#We are passing the years to df
x= range(1945,2011,5)
for i in x:  
   df_new= df_new.append(pd.DataFrame({'year':[i]}))
   df_new = df_new.reset_index(drop=True)
df_new
```

```python
#Groupby statement for every religion this way i can get the sum of religiouses by that year
a=df.groupby(['year'],as_index=False)['islam','christianity','budism','nonrelig','judaism','pop'].sum()
df_new.insert(1,'islam',a['islam'])
df_new.insert(2,'christianity',a['christianity'])
df_new.insert(3,'judaism',a['judaism'])
df_new.insert(4,'budism',a['budism'])
df_new.insert(5,'nonrelig',a['nonrelig'])
df_new.insert(6,'population',a['pop'])
```

```python
# I am exporting dataframe to an excel sheet so i can make some adjustements
df_new.to_excel(excel_writer=r"C:\Users\lenvo\Desktop\Pyhton\Sum_Religions_by_Year.xlsx")
```
