```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
url='https://medyabayim.com/services'

soup = BeautifulSoup(requests.get(url).content, "lxml")
```

```python
all_data = []
for tr in soup.select("tr:not(:has(td[colspan], th))"):
    prev = tr.find_previous("td", attrs={"colspan": True})
    tds = [td.get_text(strip=True) for td in tr.select("td")]
    all_data.append([prev.get_text(strip=True), *tds[:7]])
```

```python
df = pd.DataFrame(
    all_data,
    columns=["Kategori", "ID", "Servis", "1000 adet fiyatƒ±", "Minimum Sipari≈ü", "Maksimum Sipari≈ü","Ortalama S√ºre","A√ßƒ±klama"],
)
```
