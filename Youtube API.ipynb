```python
#function for video details like comment etc.
def get_video_details(video_id):
        url_2="https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&key="+API_KEY+"&part=statistics"
        response_2=requests.get(url_2).json()
        view_count=response_2['items'][0]['statistics']['viewCount']
        like_count=response_2['items'][0]['statistics']['likeCount']
        comment_count=response_2['items'][0]['statistics']['commentCount']
        return view_count,like_count,comment_count  
```

```python
#main function
def get_videos(df):
    #make api call 
    base_url="https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelId="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=10000&pageToken="
    page_token=""
    #for looping through pages 
    
    keep_reading=True
    while keep_reading:
       
        response=requests.get(base_url+page_token).json()
        
        
        #for loop 
        for video in response['items']:
            if video['id']['kind']=='youtube#video':
                    video_id=video['id']['videoId']
                    video_title=video['snippet']['title']
                    video_title= str(video_title).replace("&amp;","")
                    upload_date=video['snippet']['publishTime']
                    upload_date=str(upload_date).split('T')[0]

                    view_count,like_count,comment_count =get_video_details(video_id)
                
                    #save data to the dataframe
                    df=df.append({
                        'video_id':video_id,
                        'video_title':video_title,
                        'upload_date':upload_date,
                        'view_count':view_count,
                        'like_count':like_count,
                        'comment_count':comment_count
                        },ignore_index=True
                    )

        if response.get('nextPageToken'):
            page_token=response["nextPageToken"]
          
            
        else:
            keep_reading=False
  
    
    return df
```

```python
#main

#building our dataframe
df=pd.DataFrame(columns=['video_id','video_title','upload_date','view_count','like_count','comment_count'])
df=get_videos(df)
```
